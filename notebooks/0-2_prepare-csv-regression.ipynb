# Dataset Preparation for regression
**Instruction**  
The aim of this notebook is to make files for regression.  
The following code requires `PositiveDataset_public_YYMMDD.xlsx`, which we collected manually.  

```python
import pandas as pd
from rdkit import Chem
import os
```

```python
# Read dataset
df = pd.read_excel('../datasets/PositiveDataset_public_221014.xlsx', sheet_name='PT', index_col=0)
df.head()
```

```python
for i in range(df.shape[0]):
    mol = Chem.MolFromSmiles(df['SMILES'].iloc[i])
    smiles = Chem.MolToSmiles(mol)
    df['SMILES'].iloc[i] = smiles
```

```python
df.head()
```

```python
df_Tendo = df[['SMILES', 'T_endo (K)']]
df_Tendo = df_Tendo.dropna()
df_Texo = df[['SMILES', 'T_exo (K)']]
df_Texo = df_Texo.dropna()
df_Hendo = df[['SMILES', 'dH_endo (kJ/mol)']]
df_Hendo = df_Hendo.dropna()
df_Hexo = df[['SMILES', 'dH_exo (kJ/mol)']]
df_Hexo = df_Hexo.dropna()
```

```python
def getmaxmin(df, target):
    df = df.sort_values(target)
    df_min = df.drop_duplicates(subset=['SMILES'])
    df_max = df.drop_duplicates(subset=['SMILES'], keep='last')
    return df_min, df_max
```

```python
df_Tendo_min, df_Tendo_max = getmaxmin(df_Tendo, 'T_endo (K)')
df_Texo_min, df_Texo_max = getmaxmin(df_Texo, 'T_exo (K)')
df_Hendo_min, df_Hendo_max = getmaxmin(df_Hendo, 'dH_endo (kJ/mol)')
df_Hexo_min, df_Hexo_max = getmaxmin(df_Hexo, 'dH_exo (kJ/mol)')
```

```python
print('-----T_endo (K)-----')
print('Before: ', df_Tendo.shape[0])
print('After: ', df_Tendo_min.shape[0], df_Tendo_max.shape[0])

print('-----T_exo (K)-----')
print('Before: ', df_Texo.shape[0])
print('After: ', df_Texo_min.shape[0], df_Texo_max.shape[0])

print('-----dH_endo (kJ/mol)-----')
print('Before: ', df_Hendo.shape[0])
print('After: ', df_Hendo_min.shape[0], df_Hendo_max.shape[0])

print('-----dH_exo (kJ/mol)-----')
print('Before: ', df_Hexo.shape[0])
print('After: ', df_Hexo_min.shape[0], df_Hexo_max.shape[0])
```

```python
path = '../datasets/regression/'
os.makedirs(path, exist_ok=True)
df_Tendo_max.to_csv(path+'Tendo_max.csv')
df_Tendo_min.to_csv(path+'Tendo_min.csv')
df_Texo_max.to_csv(path+'Texo_max.csv')
df_Texo_min.to_csv(path+'Texo_min.csv')
df_Hendo_max.to_csv(path+'Hendo_max.csv')
df_Hendo_min.to_csv(path+'Hendo_min.csv')
df_Hexo_max.to_csv(path+'Hexo_max.csv')
df_Hexo_min.to_csv(path+'Hexo_min.csv')
```
